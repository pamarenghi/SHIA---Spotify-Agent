```python
df_muse['id_music'] = df_muse['track'] + df_muse['artist']
df_lyrics['id_music'] = df_lyrics['song'] + df_lyrics['artist']
```

```python
df = pd.merge(df_lyrics, df_muse, on='id_music')
```

# II - Sentiment analysis

```python
def analyze_mood(text):
    if not isinstance(text, str):
        text = str(text)
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    return polarity
```
